# Memória de Cálculo — Matriz de Payoff
## Estratégia de Densidade de Plantio (Milho)

**Fonte dos dados:** `harvest_summary_brazil.csv`  
**Objetivo:** Extrair os valores reais de produtividade (kg/ha) e calcular todos os critérios da Matriz de Payoff (Maximax, Maximin, Laplace e Savage).

---

## 1. Importações e Leitura dos Dados

```python
import pandas as pd
import numpy as np

df = pd.read_csv('harvest_summary_brazil.csv')

print(f'Total de registros no arquivo: {len(df):,}')
print(f'Culturas disponíveis: {sorted(df["crop_name"].unique())}')
df.head()
```

## 2. Filtragem — Apenas Milho

```python
corn = df[df['crop_name'] == 'corn'].copy()

print(f'Registros de milho: {len(corn):,}')
print(f'Safras presentes: {sorted(corn["concrete_growing_season_name"].unique())}')
print(f'\nEstatísticas de produtividade (average_yield):')
corn['average_yield'].describe().round(1)
```

## 3. Definição dos Cenários por Tercis

Como o CSV não possui uma coluna de "clima", os cenários foram derivados da **distribuição real dos yields**, dividida em tercis:

- **Pessimista** → terço inferior (yields mais baixos = condições adversas)
- **Base** → terço do meio (condições normais)
- **Otimista** → terço superior (yields mais altos = condições favoráveis)

```python
t33 = corn['average_yield'].quantile(0.33)
t66 = corn['average_yield'].quantile(0.66)

print(f'Corte P33 (pessimista/base):  {t33:.0f} kg/ha')
print(f'Corte P66 (base/otimista):    {t66:.0f} kg/ha')

pessimista = corn[corn['average_yield'] <= t33]['average_yield']
base       = corn[(corn['average_yield'] > t33) & (corn['average_yield'] <= t66)]['average_yield']
otimista   = corn[corn['average_yield'] > t66]['average_yield']

print(f'\nPessimista: n={len(pessimista):,}, faixa={pessimista.min():.0f}–{pessimista.max():.0f} kg/ha')
print(f'Base:       n={len(base):,}, faixa={base.min():.0f}–{base.max():.0f} kg/ha')
print(f'Otimista:   n={len(otimista):,}, faixa={otimista.min():.0f}–{otimista.max():.0f} kg/ha')
```

## 4. Yields Médios Reais por Cenário (Densidade Padrão)

O yield médio de cada tercil representa a **Densidade Padrão** na Matriz de Payoff.

```python
yield_padrao_pess = round(pessimista.mean())
yield_padrao_base = round(base.mean())
yield_padrao_otim = round(otimista.mean())

print('=== DENSIDADE PADRÃO (dados reais) ===')
print(f'Pessimista: {yield_padrao_pess:,} kg/ha  (média de {len(pessimista):,} talhões)')
print(f'Base:       {yield_padrao_base:,} kg/ha  (média de {len(base):,} talhões)')
print(f'Otimista:   {yield_padrao_otim:,} kg/ha  (média de {len(otimista):,} talhões)')
```

## 5. Modelagem das Alternativas de Densidade

O milho responde de forma semelhante à soja em relação à densidade, com ajustes agronômicos sobre os yields reais:

| Cenário | Reduzida −20% | Aumentada +20% |
|---|---|---|
| Pessimista | +5% (menos competição por estresse hídrico) | −10% (quebra sob estresse) |
| Base | −3% (menor teto) | +2% (leve ganho) |
| Otimista | −8% (menor teto produtivo) | +10% (máximo aproveitamento) |

```python
# Densidade Reduzida -20%
yield_reduzida_pess = round(yield_padrao_pess * 1.05 / 10) * 10
yield_reduzida_base = round(yield_padrao_base * 0.97 / 10) * 10
yield_reduzida_otim = round(yield_padrao_otim * 0.92 / 10) * 10

# Densidade Aumentada +20%
yield_aumentada_pess = round(yield_padrao_pess * 0.90 / 10) * 10
yield_aumentada_base = round(yield_padrao_base * 1.02 / 10) * 10
yield_aumentada_otim = round(yield_padrao_otim * 1.10 / 10) * 10

print('=== DENSIDADE REDUZIDA −20% ===')
print(f'Pessimista: {yield_reduzida_pess:,} kg/ha  (+5% sobre padrão)')
print(f'Base:       {yield_reduzida_base:,} kg/ha  (−3% sobre padrão)')
print(f'Otimista:   {yield_reduzida_otim:,} kg/ha  (−8% sobre padrão)')

print('\n=== DENSIDADE AUMENTADA +20% ===')
print(f'Pessimista: {yield_aumentada_pess:,} kg/ha  (−10% sobre padrão)')
print(f'Base:       {yield_aumentada_base:,} kg/ha  (+2% sobre padrão)')
print(f'Otimista:   {yield_aumentada_otim:,} kg/ha  (+10% sobre padrão)')
```

## 6. Matriz de Payoff Completa

```python
payoff = pd.DataFrame({
    'Alternativa': ['Densidade Padrão', 'Densidade Reduzida −20%', 'Densidade Aumentada +20%'],
    'Pessimista':  [yield_padrao_pess, yield_reduzida_pess, yield_aumentada_pess],
    'Base':        [yield_padrao_base, yield_reduzida_base, yield_aumentada_base],
    'Otimista':    [yield_padrao_otim, yield_reduzida_otim, yield_aumentada_otim],
}).set_index('Alternativa')

print('=== MATRIZ DE PAYOFF — MILHO (kg/ha) ===')
payoff
```

## 7. Critérios de Decisão

### 7.1 Maximax — O Otimista
Escolhe a alternativa com o **maior yield possível** (melhor dos melhores).

```python
payoff['Maximax'] = payoff.max(axis=1)

vencedor_maximax = payoff['Maximax'].idxmax()
print('Maximax por alternativa:')
print(payoff['Maximax'].to_string())
print(f'\n→ Vencedor: {vencedor_maximax} ({payoff["Maximax"].max():,} kg/ha)')
```

### 7.2 Maximin — O Conservador
Escolhe a alternativa com o **melhor yield no pior cenário** (melhor dos piores).

```python
payoff['Maximin'] = payoff[['Pessimista','Base','Otimista']].min(axis=1)

vencedor_maximin = payoff['Maximin'].idxmax()
print('Maximin por alternativa:')
print(payoff['Maximin'].to_string())
print(f'\n→ Vencedor: {vencedor_maximin} ({payoff["Maximin"].max():,} kg/ha)')
```

### 7.3 Laplace — O Neutro
Assume **probabilidade igual** para todos os cenários e escolhe a maior média.

```python
payoff['Laplace'] = payoff[['Pessimista','Base','Otimista']].mean(axis=1).round(1)

vencedor_laplace = payoff['Laplace'].idxmax()
print('Laplace por alternativa:')
print(payoff['Laplace'].to_string())
print(f'\n→ Vencedor: {vencedor_laplace} ({payoff["Laplace"].max():,.1f} kg/ha)')
```

### 7.4 Minimax Regret — Critério de Savage

Para cada cenário, calcula quanto cada alternativa "perdeu" em relação à melhor escolha possível naquele cenário.  
Escolhe a alternativa cujo **maior arrependimento seja o menor**.

```python
cenarios = ['Pessimista', 'Base', 'Otimista']

max_por_cenario = payoff[cenarios].max(axis=0)
print('Melhor yield por cenário (referência):')
print(max_por_cenario.to_string())

regret = max_por_cenario - payoff[cenarios]
print('\n=== MATRIZ DE ARREPENDIMENTO (kg/ha) ===')
print(regret.to_string())

payoff['Max Regret'] = regret.max(axis=1)
print('\nMax Regret por alternativa:')
print(payoff['Max Regret'].to_string())

vencedor_savage = payoff['Max Regret'].idxmin()
print(f'\n→ Vencedor (menor Max Regret): {vencedor_savage} ({payoff["Max Regret"].min():,} kg/ha)')
```

## 8. Resumo Final

```python
resumo = pd.DataFrame({
    'Critério': ['Maximax', 'Maximin', 'Laplace', 'Savage (Minimax Regret)'],
    'Melhor Alternativa': [
        vencedor_maximax,
        vencedor_maximin,
        vencedor_laplace,
        vencedor_savage,
    ],
    'Valor (kg/ha)': [
        payoff['Maximax'].max(),
        payoff['Maximin'].max(),
        payoff['Laplace'].max(),
        payoff['Max Regret'].min(),
    ],
    'Perfil do Decisor': [
        'Otimista — busca o maior teto produtivo',
        'Conservador — protege contra o pior clima',
        'Neutro — sem viés de cenário',
        'Pragmático — minimiza o arrependimento futuro',
    ]
})

print('=== RESUMO DE DECISÃO — MILHO ===')
resumo
```

## 9. Tabela Final para Copiar na Planilha

```python
tabela_final = payoff[['Pessimista', 'Base', 'Otimista', 'Maximax', 'Maximin', 'Laplace', 'Max Regret']]

print('=== VALORES FINAIS PARA A PLANILHA — MILHO ===')
print(tabela_final.to_string())

print('\n=== MATRIZ DE ARREPENDIMENTO COMPLETA ===')
regret_final = regret.copy()
regret_final['Max Regret'] = payoff['Max Regret']
print(regret_final.to_string())
```
